```python
import pandas as pd
import math
```

```python
matches = pd.read_csv("matches.csv")
```

```python
matches.head()
```

```python
N = matches.shape[0]
```

```python
e = 5 / 100
```

```python
n0 = 1 / (e)**2
```

```python
tamanhoAmostra = math.ceil((N * n0) / (N + n0))
```

```python
tamanhoAmostra
```

```python
amostra = matches.sample(tamanhoAmostra)
```

```python
amostra.to_csv("amostra_validar.csv", encoding="utf-8", index=False)
```
