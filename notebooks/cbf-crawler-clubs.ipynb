## Cartola Web Crawler - Clubs
Esse notebook tem como objetivo capturar os nomes dos times do Campeonato Brasileiro.
A CBF disponibiliza um web site que apresenta uma interface html muito simples e facilita o trabalho de web crawling.
Desse site estaremos extraindo os nomes dos clubes que participaram do campeonato desde a data de 2012.

```python
%pip install bs4
%pip install requests
```

```python
import requests
import re
import math
import json
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# padronizar
clubs = requests.get('https://api.cartolafc.globo.com/clubes')
json.loads(clubs.content)
```

```python
dict_clubs = {}
for year in range(2012, 2022):
    page_game = requests.get(f'https://www.cbf.com.br/futebol-brasileiro/competicoes/campeonato-brasileiro-serie-a/{year}')
    clubs = BeautifulSoup(page_game.content, 'html.parser')
    for c in clubs.find_all("span", class_="hidden-xs"):
        club = c.get_text()
        if club in dict_clubs.keys():
            dict_clubs[club]["count"] += 1
            dict_clubs[club]["year"].append(year)
        else:
            dict_clubs[club] = {}
            dict_clubs[club]["count"] = 1
            dict_clubs[club]["year"] = [year]

print(dict_clubs)
```
